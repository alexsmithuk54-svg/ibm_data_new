Print the mean age of the survey participants.


```python
## Write your code here
print("\n4. Mean age calculation:")
print("   Age column contains categorical ranges:")

# Show unique age categories
unique_ages = df['Age'].dropna().unique()
for age in sorted(unique_ages):
    print(f"   - {age}")

# Calculate approximate mean
def get_age_midpoint(age_str):
    if pd.isna(age_str):
        return None
    
    age_str = str(age_str)
    
    if 'Under 18' in age_str:
        return 16
    elif '18-24' in age_str:
        return 21
    elif '25-34' in age_str:
        return 29.5
    elif '35-44' in age_str:
        return 39.5
    elif '45-54' in age_str:
        return 49.5
    elif '55-64' in age_str:
        return 59.5
    elif '65+' in age_str or 'or older' in age_str:
        return 67.5
    else:
        return None

# Calculate mean
df['Age_numeric'] = df['Age'].apply(get_age_midpoint)
mean_age = df['Age_numeric'].mean()

print(f"\n   Approximate mean age (using category midpoints): {mean_age:.2f} years")
```
